```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMRegressor
from sklearn import neighbors
from sklearn.svm import SVR
from sklearn.ensemble import VotingClassifier
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
rf_model = RandomForestRegressor(random_state=1)
l_model = LinearRegression()
lg_model = LGBMRegressor()
k_model = neighbors.KNeighborsRegressor(n_neighbors = 5)
s_model = SVR()
```

```python
rf_model.fit(train_X,train_y)
l_model.fit(train_X,train_y)
lg_model.fit(train_X,train_y)
k_model.fit(train_X,train_y)
s_model.fit(train_X,train_y)
```

```python
rf_val_predictions = rf_model.predict(val_X)
rf_val_mse = mean_squared_error(rf_val_predictions,val_y)
```
